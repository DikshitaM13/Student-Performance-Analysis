```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#data cleaning
df.isnull().sum()
```

```python
type(df)
```

```python
df.columns
```

```python
df['Attendance(%)'] = df['Attendance (%)'].astype(int)
```

```python
df['Average Score'] = (df['Math Score'] + df['Science Score'] + df['English Score']) / 3
```

```python
df.head()
```

```python
#data analysis
df.shape[0]
```

```python
df['Result'].value_counts()
```

```python
df['Average Score'].mean()
```

```python
df.groupby('Study Hours')['Average Score'].mean()
```

```python
df.groupby('Gender')['Average Score'].mean()
```

```python
#visualization

df['Result'].value_counts().plot(kind='bar')
plt.title("Pass vs Fail")
plt.show()
```

```python
df.plot(x='Study Hours',y='Average Score',kind='scatter')
plt.title("Study Hours vs Score")
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```
